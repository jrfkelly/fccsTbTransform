```python
import pandas as pd
import pyinputplus as pyip
import glob, os, time
```

```python
#time the process
start = time.time()
```

```python
inpath = (r'C:\Users\kellyjo00\Box\FCCS Reporting\Jul-20\Input TBs')
outpath = (r'C:\Users\kellyjo00\Box\FCCS Reporting\Jul-20\CSV TBs')

all_filenames = [i for i in glob.glob(os.path.join(inpath,"*.xlsx"))]
all_filenames
```

```python
i = 0
for f in all_filenames:
    print(str(i) + ": " + f )
    i = i+1
print(' ')
file = pyip.inputNum(prompt='Which file would you like to convert?', min=0, lessThan=i)
```

```python
infile = os.path.basename(all_filenames[file])
outfile = infile.replace(".xlsx", "")
print(infile)
```

```python
df = pd.read_excel(all_filenames[file])

df.shape
```

```python
is_P1 = df['Period'] == 'P1'
is_OB = df['Movement'] == 'FCCS_OpeningBalance'

filtered_df = df[is_P1 & is_OB]
#filtered_df.shape
filtered_df.head(10)
```

```python
#set Period to "P0"
filtered_df = filtered_df.assign(Period = 'P0')
filtered_df.head(10)
```

```python
while True:
    print('')
    print('The source file is called: ' + infile)
    print('')
    outfile = pyip.inputStr(prompt='Please enter a name for the new file.', default=outfile)
    outfile = os.path.splitext(outfile)[0]+'.csv'
    print('')
    print('Your file will be saved as ' + outfile)
    response = pyip.inputYesNo(prompt='Please confirm y/n')
    if response =='yes':
        break
print(outfile)    
```

```python
#export to csv
filtered_df.to_csv(os.path.join(outpath,(os.path.splitext(outfile)[0]) + time.strftime("-%Y%m%d-%H%M") + '.csv'), index=False, encoding='utf-8-sig')
```

```python
end = time.time()
print(end - start)
```
